```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Generate synthetic data
def true_function(x):
    return np.sin(x)

np.random.seed(0)
X_train = np.random.uniform(-2*np.pi, 2*np.pi, 1000)[:, np.newaxis]
y_train = true_function(X_train) + np.random.normal(0, 0.1, size=X_train.shape)

X_val = np.random.uniform(-2*np.pi, 2*np.pi, 200)[:, np.newaxis]
y_val = true_function(X_val) + np.random.normal(0, 0.1, size=X_val.shape)

X_test = np.random.uniform(-2*np.pi, 2*np.pi, 200)[:, np.newaxis]
y_test = true_function(X_test) + np.random.normal(0, 0.1, size=X_test.shape)

# Step 2: Define and train the MLFFNN
def relu(x):
    return np.maximum(0, x)

def forward_pass(X, weights, biases):
    layer1_output = relu(np.dot(X, weights[0]) + biases[0])
    layer2_output = relu(np.dot(layer1_output, weights[1]) + biases[1])
    output = np.dot(layer2_output, weights[2]) + biases[2]
    return output

def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Initialize weights and biases
np.random.seed(0)
weights = [np.random.randn(1, 10), np.random.randn(10, 10), np.random.randn(10, 1)]
biases = [np.zeros((1, 10)), np.zeros((1, 10)), np.zeros((1, 1))]

# Hyperparameters
learning_rate = 0.001
epochs = 100

# Training loop
for epoch in range(epochs):
    # Forward pass
    y_pred = forward_pass(X_train, weights, biases)

    # Compute training loss
    train_loss = mean_squared_error(y_train, y_pred)

    # Validation
    y_val_pred = forward_pass(X_val, weights, biases)
    val_loss = mean_squared_error(y_val, y_val_pred)

    # Backpropagation
    output_error = y_pred - y_train
    layer2_error = np.dot(output_error, weights[2].T)
    layer2_delta = layer2_error * (X_train > 0).astype(float)
    layer1_delta = np.dot(layer2_delta, weights[1].T) * (X_train > 0).astype(float)

    # Update weights and biases
    weights[2] -= learning_rate * np.dot(np.transpose(layer2_output), output_error)
    weights[1] -= learning_rate * np.dot(np.transpose(layer1_output), layer2_delta)
    weights[0] -= learning_rate * np.dot(np.transpose(X_train), layer1_delta)
    biases[2] -= learning_rate * np.sum(output_error, axis=0, keepdims=True)
    biases[1] -= learning_rate * np.sum(layer2_delta, axis=0, keepdims=True)
    biases[0] -= learning_rate * np.sum(layer1_delta, axis=0, keepdims=True)

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Training Loss: {train_loss}, Validation Loss: {val_loss}")

# Step 3: Evaluate the model using test data
y_test_pred = forward_pass(X_test, weights, biases)
test_loss = mean_squared_error(y_test, y_test_pred)
print(f"Test Loss: {test_loss}")

# Plot the approximated function surface with training data points
x_test = np.linspace(-2*np.pi, 2*np.pi, 100).reshape(-1, 1)
y_pred = forward_pass(x_test, weights, biases)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x_test, true_function(x_test), color='r', label='True Function')
ax.scatter(X_train, y_train, color='b', alpha=0.3, label='Training Data')
ax.plot(x_test, y_pred, color='g', label='Approximated Function')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()
plt.show()
```
