```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Generate synthetic data (replace with your actual data)
X_train = np.random.rand(100, 2)  # Training input data
y_train = np.sin(X_train[:, 0]) + np.cos(X_train[:, 1])  # Example function

# Split data into training, validation, and test sets
val_size = 10  # Number of samples for validation
test_size = 10  # Number of samples for testing
X_val = X_train[-val_size:]
y_val = y_train[-val_size:]
X_train = X_train[:-val_size]
y_train = y_train[:-val_size]

# Build the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)  # Regression output
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=100, verbose=0, validation_data=(X_val, y_val))

# Create a grid for surface plot
input1_vals = np.linspace(0, 1, 50)
input2_vals = np.linspace(0, 1, 50)
input1_grid, input2_grid = np.meshgrid(input1_vals, input2_vals)
input_grid = np.column_stack((input1_grid.ravel(), input2_grid.ravel()))

# Predict output for the grid
predicted_output = model.predict(input_grid).reshape(input1_grid.shape)

# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(input1_grid, input2_grid, predicted_output, cmap='viridis', alpha=0.7, label='Approximated Function')
ax.set_xlabel('Input 1')
ax.set_ylabel('Input 2')
ax.set_zlabel('Output')
ax.set_title('Approximated Function Surface')

# Overlay training data points
ax.scatter(X_train[:, 0], X_train[:, 1], y_train, color='red', label='Training Data')

# Add legends
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def relu(x):
    return np.maximum(0, x)

# Generate synthetic data (replace with your actual data)
X_train = np.random.rand(100, 2)  # Training input data
y_train = np.sin(X_train[:, 0]) + np.cos(X_train[:, 1])  # Example function

# Split data into training, validation, and test sets
val_size = 10  # Number of samples for validation
test_size = 10  # Number of samples for testing
X_val = X_train[-val_size:]
y_val = y_train[-val_size:]
X_train = X_train[:-val_size]
y_train = y_train[:-val_size]

# Initialize weights and biases
np.random.seed(0)
weights = [np.random.randn(2, 64), np.random.randn(64, 32), np.random.randn(32, 1)]
biases = [np.zeros((1, 64)), np.zeros((1, 32)), np.zeros((1, 1))]

# Hyperparameters
learning_rate = 0.001
epochs = 100

# Training loop
for epoch in range(epochs):
    # Forward pass
    layer1_output = relu(np.dot(X_train, weights[0]) + biases[0])
    layer2_output = relu(np.dot(layer1_output, weights[1]) + biases[1])
    y_pred = np.dot(layer2_output, weights[2]) + biases[2]
    
    # Compute loss
    loss = np.mean((y_train - y_pred) ** 2)
    
    # Backpropagation
    output_error = y_pred - y_train
    layer2_error = output_error
    layer2_delta = layer2_error * (layer2_output > 0)  # ReLU derivative
    layer1_error = np.dot(layer2_delta, weights[2].T)
    layer1_delta = layer1_error * (layer1_output > 0)  # ReLU derivative
    
    # Update weights and biases
    weights[2] -= learning_rate * np.dot(layer2_output.T, output_error)
    weights[1] -= learning_rate * np.dot(layer1_output.T, layer2_delta)
    weights[0] -= learning_rate * np.dot(X_train.T, layer1_delta)
    biases[2] -= learning_rate * np.sum(output_error, axis=0, keepdims=True)
    biases[1] -= learning_rate * np.sum(layer2_delta, axis=0, keepdims=True)
    biases[0] -= learning_rate * np.sum(layer1_delta, axis=0, keepdims=True)

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

# Create a grid for surface plot
input1_vals = np.linspace(0, 1, 50)
input2_vals = np.linspace(0, 1, 50)
input1_grid, input2_grid = np.meshgrid(input1_vals, input2_vals)
input_grid = np.column_stack((input1_grid.ravel(), input2_grid.ravel()))

# Predict output for the grid
layer1_output = relu(np.dot(input_grid, weights[0]) + biases[0])
layer2_output = relu(np.dot(layer1_output, weights[1]) + biases[1])
predicted_output = np.dot(layer2_output, weights[2]) + biases[2]
predicted_output = predicted_output.reshape(input1_grid.shape)

# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(input1_grid, input2_grid, predicted_output, cmap='viridis', alpha=0.7, label='Approximated Function')
ax.set_xlabel('Input 1')
ax.set_ylabel('Input 2')
ax.set_zlabel('Output')
ax.set_title('Approximated Function Surface')

# Overlay training data points
ax.scatter(X_train[:, 0], X_train[:, 1], y_train, color='red', label='Training Data')

# Add legends
ax.legend()

plt.show()
```
