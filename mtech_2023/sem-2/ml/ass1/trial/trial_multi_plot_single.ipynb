```python
import matplotlib.pyplot as plt

# Example data
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]

x2 = [2, 3, 4, 5, 6]
y2 = [5, 6, 8, 10, 13]

# Plot the first set of points (blue)
plt.scatter(x1, y1, color='blue', label='Set 1')

# Plot the second set of points (red)
plt.scatter(x2, y2, color='red', label='Set 2')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Scatter Plots')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
