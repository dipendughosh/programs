```python
import csv
import random

# Function to generate unique random coordinates within a given range
def generate_coordinates(count, lower_bound, upper_bound):
    coordinates = set()
    while len(coordinates) < count:
        x = random.randint(lower_bound, upper_bound)
        y = random.randint(lower_bound, upper_bound)
        coordinates.add((x, y))
    return coordinates

# Initialize location data structure to store multiple sets
location_data = []

# Generate 100 sets of anchors and corresponding node locations
for i in range(100):
    # Generate anchor locations (3 coordinates per set)
    anchor_set = generate_coordinates(3, 0, 100)
    print(f"Anchor Set {i+1}: {anchor_set}")
    
    # Generate node locations (50 coordinates per set)
    node_set = generate_coordinates(50, 0, 100)
    print(f"Node Set {i+1}: {node_set}")
    
    # Append each pair of anchor and node sets to location_data
    location_data.append({
        "Anchor Set": list(anchor_set),
        "Node Set": list(node_set)
    })

# Check the location_data after all iterations
print(f"\nLocation Data after 100 sets: {location_data[:3]}...")  # Print only first 3 for brevity

# Define the filename for the CSV output
filename = 'dipendu_data/true_locations_100.csv'

# Write location_data to CSV with each row representing one anchor-node pair
with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # Write the header row
    header = [f"Anchor_{i+1}" for i in range(3)] + [f"Node_{j+1}" for j in range(50)]
    writer.writerow(header)
    print(f"Header written to CSV: {header}")
    
    # Write each anchor-node pair in a single row
    for idx, data in enumerate(location_data):
        # Format anchor and node coordinates as (x, y) strings
        row = [f"({x}, {y})" for (x, y) in data["Anchor Set"]] + \
              [f"({x}, {y})" for (x, y) in data["Node Set"]]
        writer.writerow(row)
        if idx < 3:  # Print first 3 rows for verification
            print(f"Row {idx+1} written to CSV: {row}")

print(f"Data has been written to {filename}")
```

```python










# Function to calculate Euclidean distance
def euclidean_distance(coord1, coord2):
    return np.sqrt((coord1[0] - coord2[0])**2 + (coord1[1] - coord2[1])**2)

# Function to calculate RMSE for a candidate (i, j)
def calculate_rmse(candidate, anchors, ranges):
    estimated_distances = [euclidean_distance(candidate, anchor) for anchor in anchors]
    rmse = np.sqrt(np.mean([(estimated_distances[k] - ranges[k])**2 for k in range(3)]))
    return rmse

# Generate random row (A) and node (B)
A = random.randint(0, 100)
print(f"{A}")
B = random.randint(0, 50)
print(f"{B}")

# Extract anchor and node coordinates from the chosen row
anchors = [eval(true_locations.iloc[A, i]) for i in range(3)]  # Convert strings to tuples
print(f"{anchors}")
node = eval(true_locations.iloc[A, 4 + B])                 # Convert node coordinates to tuple
print(f"{node}")

# Extract the range values for the chosen row and node from all files
pure_range = eval(pure_ranges.iloc[A, 4 + B])                   # Distances for chosen node
print(f"{pure_range}")
noisy_range_05 = eval(noisy_ranges_05.iloc[A, 4 + B])
print(f"{noisy_range_05}")
noisy_range_1 = eval(noisy_ranges_1.iloc[A, 4 + B])
print(f"{noisy_range_1}")
noisy_range_2 = eval(noisy_ranges_2.iloc[A, 4 + B])
print(f"{noisy_range_2}")

# Prepare matrices to store cost values
cost_matrix_pure = np.zeros((100, 100))
cost_matrix_noisy_05 = np.zeros((100, 100))
cost_matrix_noisy_1 = np.zeros((100, 100))
cost_matrix_noisy_2 = np.zeros((100, 100))

# Evaluate cost function for each (i, j) in a 100x100 grid
for i in range(100):
    for j in range(100):
        candidate_location = (i, j)
        # Calculate and store RMSE-based cost values in matrices
        cost_matrix_pure[i, j] = calculate_rmse(candidate_location, anchors, pure_range)
        cost_matrix_noisy_05[i, j] = calculate_rmse(candidate_location, anchors, noisy_range_05)
        cost_matrix_noisy_1[i, j] = calculate_rmse(candidate_location, anchors, noisy_range_1)
        cost_matrix_noisy_2[i, j] = calculate_rmse(candidate_location, anchors, noisy_range_2)

# The resulting matrices contain RMSE cost values for each (i, j) in the grid
# print("Cost matrix for pure_ranges.csv:\n", cost_matrix_pure)
# print("Cost matrix for noisy_ranges_05.csv:\n", cost_matrix_noisy_05)
# print("Cost matrix for noisy_ranges_1.csv:\n", cost_matrix_noisy_1)
# print("Cost matrix for noisy_ranges_2.csv:\n", cost_matrix_noisy_2)
```
